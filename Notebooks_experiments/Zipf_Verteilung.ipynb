```python
from scipy.stats import zipfian
import matplotlib.pyplot as plt
import matplotlib


font = {#'family' : 'normal',
        #'weight' : 'normal',
        'size'   : 16}

matplotlib.rc('font', **font)
fig, ax = plt.subplots(1, 1)
a, n = 1, 11
mean, var, skew, kurt = zipfian.stats(a, n, moments='mvsk')

x = np.arange(zipfian.ppf(0.01, a, n),
              zipfian.ppf(0.99, a, n))
ax.plot(x, zipfian.pmf(x, a, n), 
        #label='zipfian pmf',
        color='r', marker="x")
#ax.vlines(x, 0, zipfian.pmf(x, a, n), colors='b', lw=5, alpha=0.5)
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
rv = zipfian(a, n)
#ax.vlines(x, 0, rv.pmf(x), colors='k', linestyles='-', lw=1,
#        label='frozen pmf')
ax.bar(x, rv.pmf(x))
ax.legend(loc='best', frameon=False)
ax.set_xlabel("Group $G_i$")
ax.set_ylabel("Probability to sample $G_i$")
ax.set_xticks([i for i in range(1, n)])
ax.set_ylim((0, 0.5))
ax.set_yticks([0.1*i for i in range(6)])
plt.savefig("zipfian_dist_example1.pdf",  bbox_inches='tight')
```

## Plank Distribution

Not able to specify how many values we have to sample from 

```python
from scipy.stats import planck
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1, 1)
lambda_ = 0.1
mean, var, skew, kurt = planck.stats(lambda_, moments='mvsk')
x = np.arange(planck.ppf(0.01, lambda_),
              planck.ppf(0.99, lambda_))
ax.plot(x, planck.pmf(x, lambda_), 'bo', ms=8, label='planck pmf')
ax.vlines(x, 0, planck.pmf(x, lambda_), colors='b', lw=5, alpha=0.5)
r = planck.rvs(lambda_, size=100)
print(np.unique(r, return_counts=True))
```

```python
from scipy.stats import poisson
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1, 1)
mu = 1.0
mean, var, skew, kurt = poisson.stats(mu, moments='mvsk')
x = np.arange(poisson.ppf(0.01, mu),
              poisson.ppf(0.99, mu))
ax.plot(x, poisson.pmf(x, mu), 'bo', ms=8, label='poisson pmf')
ax.vlines(x, 0, poisson.pmf(x, mu), colors='b', lw=5, alpha=0.5)
rv = poisson(mu)
ax.vlines(x, 0, rv.pmf(x), colors='k', linestyles='-', lw=1,
        label='frozen pmf')
ax.legend(loc='best', frameon=False)
plt.show()
r = poisson.rvs(mu, size=1000)
print(np.unique(r, return_counts=True))
```

```python
from scipy.stats import norm
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1, 1)
mean, var, skew, kurt = norm.stats(moments='mvsk')
x = np.linspace(norm.ppf(0.01),
                norm.ppf(0.99), 100)
ax.plot(x, norm.pdf(x),
       'r-', lw=5, alpha=0.6, label='norm pdf')
rv = norm()
ax.plot(x, rv.pdf(x), 'k-', lw=2, label='frozen pdf')
r = norm.rvs(size=1000)
ax.hist(r, density=True, histtype='stepfilled', alpha=0.2)
ax.legend(loc='best', frameon=False)
plt.show()
```

```python
from anytree import Node
from scipy.stats import zipfian
from scipy.stats import zipf
from collections import Counter
from anytree import RenderTree
n_samples = 1000
n_classes = 100

childs_per_level = 3
root = Node("root")
root.n = n_samples
root.c = n_classes

current_nodes = [root]
level = 0
max_level = 3

while level < max_level:
    leaves = root.leaves
    for node in leaves:
        counter = zipfian.rvs(a=1.25, size=node.n, n=childs_per_level)
        counter = list(Counter(counter).values())
        counter = sorted(counter)

        class_counter = zipfian.rvs(a=1, size=int(node.c * 1.5) , n=childs_per_level)
        class_counter = list(Counter(class_counter).values())
        class_counter = sorted(class_counter)
        count = 0
        for i in range(childs_per_level):
            count+=1
            new_node = Node(f"level={level};id={count}", parent=node)
            new_node.n = counter[i]
            new_node.c = class_counter[i]
            current_nodes.append(new_node)
            
        
    level += 1
    
    #print(sum(counter.values()))
print(RenderTree(root))
```

```python
n_samples = sum(occurences)
max_occurence = max(occurences)
max_index = occurences.index(max_occurence)
new_occurences = occurences.copy()

median = int(np.median(occurences))
average = sum(occurences)/len(occurences)
median_or_average = int(average) if median == 1 else median

def _eq_div(N, i):
    """
    Divide N into i buckets while preserving the remainder to the buckerts as well.
    :return: list of length i
    """
    return [] if i <= 0 else [N // i + 1] * (N % i) + [N // i] * (i - N % i)


n_max_classes = 1
n_max_classes = len([x for x in occurences if x > average])

print(n_max_classes)
print(f"median is: {median}")
print(f"average is: {average}")
# for each class above average, we run the following procedure
print(f"occurences before modification: {occurences}")
print(f"gini score before modification: {gini_before}")
print(f"occurences above {median_or_average} are: {[x for x in occurences if x > median_or_average]}")

median_or_average = int(average) if median == 1 else int(median)

for i in range(n_max_classes):
    max_occurence = max(occurences)
    max_index = occurences.index(max_occurence)
    new_occurences = occurences.copy()
    print(f"occurences before modification: {occurences}")

    #if len(new_occurences) < max_occurence:
    new_occurences[max_index] = median_or_average   

    # equal division of max - average
    rest = _eq_div((max_occurence - median_or_average), len(new_occurences)-1)

    # insert 0 at max position -> We do not want to add something to max.
    rest.insert(max_index, 0)

    for i, r in enumerate(rest):
        new_occurences[i] +=r
    occurences = new_occurences

    
occ_dic = {i: [x] for i, x in enumerate(new_occurences)}
class_frequencies = np.array(list(occ_dic.values()))
my_index = cm.Index()
gini_after = my_index.gini(class_frequencies)
print(f"occurences after modification: {new_occurences}")
print(f"gini score after modification: {gini_after}")
```
