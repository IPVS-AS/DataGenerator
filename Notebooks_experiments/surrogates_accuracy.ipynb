```python
import pandas as pd
import numpy as np
import os
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
merged_df = pd.DataFrame()
for imb in ['very_low', 'low', 'normal', 'high', 'very_high']:
    file_name = f'../surrogates/imbalance_degree/{imb}/result/surrogates.csv'

    surrogates_df = pd.read_csv(file_name)
    file_name = f'../surrogates/imbalance_degree/{imb}/result/predictions.csv'
    surrogate_groups = surrogates_df["surrogate"].values
    pred_df = pd.read_csv(file_name)
    pred_df["imb"] = imb
    
    pred_df["surrogate"] = pred_df.apply(lambda x: x["group"] in surrogate_groups, axis=1)
    print(pred_df.shape)
    merged_df = pd.concat([merged_df, pred_df])
merged_df
```

```python
for imb in merged_df["imb"].unique():
    print("-----------------------------------")
    print(f"imbalance Degree: {imb}")
    imb_df = merged_df[merged_df["imb"] == imb]
    #print(f"surrogate samples: {len(imb_df)}")
    surrogates = imb_df[imb_df["surrogate"]]
    print(f"#surrogate samples: {len(surrogates)}")
    primary_sets = imb_df[~imb_df["surrogate"]]
    print(f"#primary samples: {len(primary_sets)}")
```

```python
result = {"Method": [], "imbalance": [], "primary A@1": [], "surrogate A@1": [],
          "primary Avg A@e": [], "surrogate Avg A@e": [], "imbalance (#p/#s)": [] 
         # "#primary samples": [], "#surrogate samples": []
         }

for imb in merged_df["imb"].unique():
    print("-----------------------------------")
    print(f"imbalance Degree: {imb}")
    imb_df = merged_df[merged_df["imb"] == imb]
    for method in merged_df["Method"].unique():
        if (method == "RF") or (method =="CPI"):
            continue
        method_df = imb_df[imb_df["Method"] == method]
        #print(f"surrogate samples: {len(imb_df)}")
        
        primary_acc = 0
        surrogate_acc = 0
        
        primary_sets = method_df[~method_df["surrogate"]]
        n_primaries = len(primary_sets)
        surrogates = method_df[method_df["surrogate"]]
        n_surrogates = len(surrogates)
        
        for e in range(1, 10+1):
            n_surroages_correct = len(surrogates[(surrogates["correct_position"] <= e) & (surrogates["correct_position"] > 0)])
            print(f"#surrogate samples: {len(surrogates)}")
            surrogate_acc += round(100*n_surroages_correct/n_surrogates)
            print(f"surrogate accuracy: {surrogate_acc}")

            
            n_primaries_correct = len(primary_sets[(primary_sets["correct_position"] <= e) & (primary_sets["correct_position"] > 0)])
            print(f"#primary samples: {len(primary_sets)}")
            primary_acc += round(100*n_primaries_correct/n_primaries)
            
            
            print(f"primary accuracy: {round(100*n_primaries_correct/n_primaries)}")
            if e ==1:
                result["primary A@1"].append(primary_acc)
                result["surrogate A@1"].append(surrogate_acc)
        result["Method"].append(method)
        result["imbalance"].append(imb)
        
        result["primary Avg A@e"].append(primary_acc/10)
        result["surrogate Avg A@e"].append(surrogate_acc/10)
        
        result["imbalance (#p/#s)"].append(f"{imb} ({n_primaries}/{n_surrogates})")
        
result = pd.DataFrame(result)
result_a_1 = pd.melt(result, id_vars=["Method", "imbalance (#p/#s)"], value_vars=[ "primary A@1", "surrogate A@1"])
result_avg = pd.melt(result, id_vars=["Method", "imbalance (#p/#s)"], value_vars=[ "primary Avg A@e", "surrogate Avg A@e",])
print(result)
result_a_1 = result_a_1.rename({"variable": "primary/surrogate", "value": "A@1"}, axis='columns')
result_a_1["primary/surrogate"] = result_a_1["primary/surrogate"].apply(lambda x: x.split(' ')[0])
result_avg = result_avg.rename({"variable": "primary/surrogate", "value": "Avg A@e"}, axis='columns')
result_avg["primary/surrogate"] = result_avg["primary/surrogate"].apply(lambda x: x.split(' ')[0])
result_a_1
```

```python
methods_ordered = ["SPH+CPI", "SPH", "RF+B", 
                   #"CPI"
                  ]

def write_values_lineplot(df, ax):
    palette = sns.color_palette()[0:5]
    color_per_label = {x:c for x,c in zip(methods_ordered, palette)}

    for item, color in zip(df.groupby('method'),palette):
        #item[1] is a grouped data frame
        for x,y,method in item[1][['R_e','A@e','method']].values:
            y = np.mean(y)
            y_position = round(y)
            x_position = x
            if method == "SPH+CPI":
                y_position = y_position +3
                x_position = x_position - 0.25
            else:
                y_position = y_position - 3
                if method == "SPH":
                    x_position = x + 0.25
                    if y_position == 33:
                        y_position = y_position + 7
                        print(y_position)
                if method == "CPI":
                    y_position = y_position - 3
                if method == "RF+B":
                    y_position = y_position - 1
                    
                
            ax.text(x_position,y_position,f'{y:.0f}%', color=color_per_label[method])
            
best_all_methods = result_a_1
#sns.set_style("darkgrid")

ax = sns.lineplot(x="imbalance (#p/#s)", y="A@1", data=best_all_methods, hue="Method", hue_order=methods_ordered, style="primary/surrogate")
#write_values_lineplot(best_all_methods, ax)

ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
#ax.set_xlabel(r'Recommendation lists $R_e$ with different lengths $e$')
ax.set_ylim(0, 100)
ax.set_ylabel('A@1')

#plt.legend(bbox_to_anchor=(1, -0.3), loc='lower right', ncol=5)
plt.tight_layout()
#plt.savefig(f"accuracy_{imb}_no_cpi.pdf")
#plt.close()

sns.catplot(x="Method", y="A@1", hue="primary/surrogate",
            col="imbalance (#p/#s)", aspect=.7,
            kind="bar", data=result_a_1)
plt.show()
```
