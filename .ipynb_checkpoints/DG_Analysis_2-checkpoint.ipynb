# Loading and preparing the Data

```python
def adjust_group(row):
    if not pd.isna(row["group"]):
        return row["group"]
    
    if pd.isna(row["Unnamed: 0.1"]):
        return row["Unnamed: 0"]
    
    if pd.isna(row["Unnamed: 0"]):
        return row["Unnamed: 0.1"]
    
    else:
        return row["Unnamed: 0.1"]

stats_df = pd.read_csv("evaluation/stats.csv", sep=';', decimal=',', index_col=None)
preds = pd.read_csv("evaluation/predictions.csv", sep=';', decimal=',', index_col=None)
print(preds.head())
preds["group"] = preds.apply(lambda row: adjust_group(row), axis=1)
#preds = preds.drop(["Unnamed: 0", "Unnamed: 0.1"], axis=1)
for column in ["accuracy", "F1", "prec", "recall"]:
    preds[column] = preds[column].astype(float)
#stats_df = stats_df.drop(["Unnamed: 0.1", "Unnamed: 0.1.1"], axis=1)
stats_df = stats_df[(stats_df["n"] == 1000) &
                    (stats_df["c"].isin( [10,20,30,40,50]))]
preds = preds[(preds["n"] == 1000)&
    (preds["c"].isin( [10,20,30,40,50]))]

#complexity_names = {'Fishers DR (C)': 'Fishers DR (F1)',
#       'Class Overlap (C)': 'Class Overlap (F2)', 'Border Points (C)': 'Border Points (N1)', 
#                    'Inter/Intra Class Dist (C)': 'Inter/Intra Class Dist (N2)', 
#                    'NN Error (C)': 'NN Error (N3)'}
#stats_df.columns.rename(complexity_names)
stats_df
```

```python
from pymfe.mfe import MFE
# Complexity metrics to measure
metric_mapping = {'c1': "Class entropy",
                  "c2": "Imbalance Degree",
                  "f1v.mean": "Fishers DR",
                  "f2.mean": "Class Overlap",
                  "vdu": "Dunn Index",
                  "n1": "Border Points",
                  "n2.mean": "Inter/Intra Class Dist",
                  "sil": "SIL",
                  "n3.mean": "NN Error",
                  "ch": "CHI",
                  "l3.mean": "NL LC",
                  "l1.mean": "Sum Error LP",
                  "t1.mean": "Hyperspheres",
                  "density": "Density",
                  "int": "INT",
                    "pb": "PB",
                  "vdb": "VDB"
                  }
```

```python
def extract_mfes(X, y, meta_feature_set=None, summary=["mean"], groups=["all"]):
    mfe = MFE(groups=groups, 
#               features=meta_feature_set,
              summary=summary)
    mfe.fit(X, y)
    ft = mfe.extract()
    return ft
```

```python
def extract_complexity_measures(X,y):
    #complexity_metrics = ["f1v", "f2", "n1", "n2", "c1", "c2", "n3","l3","l1","t1","density"]
    #CVI_s = ["sil", "ch", "vdu", "int", "pb", "vdb"]
    #####################################
    ### Complexity metrics  from PyMFE ###
    ft = extract_mfes(X, y, groups=["complexity", "clustering"])
    #for metric, value in zip(ft[0], ft[1]):
        #print(f"{metric_mapping[metric]} (C): {value}")
        #print(f"{metric}: {value}")
    return dict(zip(ft[0], ft[1]))
```

```python
from sklearn.datasets import make_classification
from DataGenerator import ImbalanceGenerator
from Hierarchy import EngineTaxonomy
generator = ImbalanceGenerator(n=1000, c=50, n_features=50, group_imbalance='medium', cls_imbalance='medium', gs=0, root=EngineTaxonomy().create_taxonomy(), features_remove_percent=0)
df = generator.generate_data_with_product_hierarchy()
#extract_complexity_measures(X,y)
compl_metrics = {}
for group in df["group"].unique():
    group_df = df[df["group"] == group]
    X = group_df[[f"F{i}" for i in range(50)]].to_numpy()
    y = group_df["target"].to_numpy()
    group_metrics = extract_complexity_measures(X,y)
    for metric,value in group_metrics.items():
        if metric in compl_metrics:
            compl_metrics[metric].append(value)
        else:
            compl_metrics[metric] = [value]
```
